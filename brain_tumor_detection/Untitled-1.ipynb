```python

import numpy as np  # to deal with matrix and numbers
import cv2  # to read files and img
import os  # to deal with oprating system
from PIL import Image  # to resize img and turn bgr to rbg
from sklearn.model_selection import train_test_split  # to train the model
#import tensorflow as tf
#from tensorflow import keras
# to normalize data --- to work we should import tenserflow and from//import keras
from keras.utils import normalize
# use to create a nural netowrk model ##انشاء نموذج شبكه عصبية
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Activation, Dropout, Flatten, Dense
dataset = []
res = []
img_size = 64

# this to simplify acsess to pic dir
image_directory = "dataset/"
# this command to list our pic in list
no_tumor_image = os.listdir(image_directory + "no/")
yes_tumor_image = os.listdir(image_directory + "yes/")
# this arraies to put data and get results

# this loob to numerate img and cheek  if all img where end with jpg and read img and turn img from bgr to rbg and also resize it to 64*64
for i,  image_name in enumerate(no_tumor_image):
    if (image_name.split(".")[1] == "jpg"):
        # now we will convert this imgs to rbg to make process on it and open cv read rbg format only
        image = cv2.imread(image_directory + "no/"+image_name)
        # this to convert from bgr to rbg
        image = Image.fromarray(image, "RGB")
        image = image.resize((img_size, img_size))
        # this to append imgs to one array (dataset)
        dataset.append(np.array(image))
        res.append(0)


for i,  image_name in enumerate(yes_tumor_image):
    if (image_name.split(".")[1] == "jpg"):
        # now we will convert this imgs to rbg to make process on it and open cv read rbg format only
        image = cv2.imread(image_directory + "yes/"+image_name)
        # this to convert from bgr to rbg
        image = Image.fromarray(image, "RGB")
        image = image.resize((img_size, img_size))  # this to resize to 64*64
        dataset.append(np.array(image))
        res.append(1)
        # now we made a large array that contain no and yes tumor and turned to rbg and resized to one size

# we will turn the  usuall array to np array to let us do functions inside numpy
dataset = np.array(dataset)
res = np.array(res)

x_train, x_test, y_train, y_test = train_test_split(dataset, res, test_size=0.2, random_state=0)

# why only x train and x test .. y train ytest ??
x_train = normalize(x_train, axis=1)
x_test = normalize(x_test, axis=1)

########## building model###########

model = Sequential()  # here we made an empty nural network object
# 3 refare to what?
model.add(Conv2D(32, (3, 3), input_shape=(img_size, img_size, 3)))
model.add(Activation("relu"))  # activion fun
model.add(MaxPooling2D(pool_size=(2, 2)))


model.add(Conv2D(32, (3, 3), kernel_initializer="he_uniform"))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2, 2)))


model.add(Conv2D(64, (3, 3), kernel_initializer="he_uniform"))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2, 2)))


model.add(Flatten())
model.add(Dense(64))
model.add(Activation("relu"))
model.add(Dropout(0.5))
model.add(Dense(1))  # what is 2
model.add(Activation("sigmoid"))

model.compile(loss="binary_crossentropy",
              optimizer="adam", metrics=["accuracy"])
model.fit(x_train, y_train, batch_size=16, verbose=1, epochs=100,
          validation_data=(x_test, y_test), shuffle=False)
model.save("BrainTumor7.h5")
```
